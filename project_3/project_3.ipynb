1. Pillbox

Nothing too crazy here. I think we did this in Prof. Shylnov's class actually. Our pillbox is a cylinder with radius $a$ and length $h$. Combine Maxwell's equations

$$
\begin{align*}
    \nabla^2 \vec E &= - \nabla \times \nabla \times \vec E
\\  &= \nabla \times {\partial \vec B \over \partial t}
     =  {\partial  \over \partial t} \left(\nabla \times \vec B\right)
\\  &=  \mu_0 {\partial \vec J \over \partial t} + \mu_0 \epsilon_0 { \partial^2 E \over \partial t^2 }.
\end{align*}
$$
No current in the cavity space, so this leaves us with
$$
\nabla^2 \vec E = { 1 \over c^2} { \partial^2 E \over \partial t^2 }.
$$

Now I can copy/paste all my old notes! $\def\d#1#2{\frac{\partial #1}{\partial #2}}$ 
Our final solution should be entirely in the $z$ direction, so in cylindrical coordinates we have

$$
    \d{^2E_z}{t^2} = c^2 \left( \d{^2 E_z}{r^2} + {1\over r} \d{E_z}{r} + {1 \over r^2} \d{^2E_z}{\phi^2} + \d{^2E_z}{z^2}\right),
$$
and we assume the periodic solution
$$
\begin{gathered}
\\  \vec E = e^{\pm i \omega t} u(r,\phi,t),
\\  \nabla^2 u + {\omega^2 \over c^2} u = 0
    \quad \rightarrow \quad
    {\omega \over c} = k,
\\  \d{^2 u}{r^2} + {1\over r} \d{u}{r} + {1 \over r^2} \d{^2u}{\phi^2} + \d{^2u}{z^2} + k^2 u = 0.
\end{gathered}
$$

The cavity boundary conditions really just boil down to the perfect conductor $\lambda_L=0$, alongside enforcing  $\nabla \cdot \vec E = 0$. I think these are all of them:
$$
\begin{gathered}
    E_z(a,\phi,z,t)=0,
\\  u  (a,\phi,z)=0,
\\  u(0,\phi,z) \text{ finite},
\\  u(r,\phi,z) = u(r,\phi+2\pi,z),
\\  \d{E_z}{z} = 0, \quad z = 0,h.
\end{gathered}
$$
I think we can find the other two components $E_r$ and $E_\phi$ at the end.

Insert separable solution
$$
\begin{gathered}
    E_z = R \Phi Z,
\\    \frac{R'' + {1\over r}R'}{R} 
    + {1\over r^2} \underbrace{\frac{\Phi''}{\Phi}}_{-n^2} 
    + \underbrace{\frac{Z''}{Z}}_{-\mathscr K^2}
    + k^2
    = 0,
\\  \Phi = A \cos (n \phi) + B \sin (n\phi),
\\  Z = c_1 \cos(\mathscr K z) + c_2 \sin(\mathscr K z).
\\  \mathscr K h = \pi m,
\\  Z = \cos \left( \pi m z \over h\right),
\end{gathered}
$$
These come from the periodic and Neumann boundary conditions, as well as the nonzero $z=0$ term.
$$
    R'' + {1 \over r} R' + \left[k^2 - \mathscr K^2-{n^2\over r^2}\right] R = 0.
$$
To make this look more like a bessel differential equation, we use
$$
      \gamma_{m n}^2 
    = k^2 - \mathscr K^2 
    = {\omega_{mn}^2 \over c^2} 
    - \left( \pi m \over h \right)^2 
    > 0
$$
with the greater than relation required for real solutions. This gives us
$$
\begin{gathered}
\\    R(r) 
    = A J_n (\gamma_{mn}r)
    + B N_n (\gamma_{mn}r),
\\  R(0) \text{ finite } \rightarrow B = 0,
\\  \gamma_{mn} a = \alpha_{ns} \rightarrow J_n(\alpha_{ns})=0,
\\  \gamma_{mn}^2 = \left( \alpha_{ns} \over a\right)^2,
\\[0.25em]
    \omega_{mns}^2 = \left( \alpha_{ns}c \over a \right)^2 + \left( \pi m c \over h \right)^2.
\end{gathered}
$$
With this choice we have no electric field at the edges, $\text{TM}_{nsm}$
- Azimuthal n
- Radial s
- Axial m
$$
      E_z(r, \phi,z,t) 
    = J_n \left( \alpha_{ns} r \over a \right)
      \left[A \cos(n\phi) + B \sin(n\phi) \right]
      \cos \left( \pi m z \over h \right)
      e^{\pm i \omega_{nsm}t},
$$
where $A$, $B$, and the time dependence sign are determined by how the mode is excited. For $\text{TM}_{010}$, this becomes
$$
      E_z(r, \phi,z,t) 
    = A J_0 \left( \alpha_{01} r \over a \right)
      e^{\pm i \omega_{010}t}.
$$

We can extract $\vec B$ using

$$
% \begin{align*}
\nabla \times \vec B = {1 \over c^2} {\partial \vec E \over \partial t}.
% \end{align*}
$$

I think we are free to choose $B_z = 0$ since accelerators use the $\text{TM}$ mode, so we'll assume this to make things easier. This reduces the curl down to

$$
% \begin{gathered}
      \nabla \times \vec B 
    = - \d{B_\phi}{z} \hat r
    + \d{B_r}{z} \hat \phi,
% \end{gathered}
$$

Having no $B_z$ term also allows us to solve for the other components of the electric field.

$$
\begin{gathered}
      \left( \nabla \times \vec E\right)_z 
    = -\d{B_z}{t} 
    = 0,
\\    \d{(r E_\phi)}{r} = \d{E_r}{\phi}

\end{gathered}
$$

$$
\begin{gathered}
      \nabla \cdot \vec E 
    = {1 \over r} \d{r E_r}{dr} + {1 \over r} \d{E_\phi}{\phi} + \d{E_z}{z} 
    = 0,
\\    \d{E_z}{z} 
    = - {1 \over r} \d{(r E_r)}{r} - {1 \over r} \d{E_\phi}{\phi}
%     = - {2 \over r} \d{E_\phi}{\phi},

2. Longitudinal Dynamics.

I got this one started in class using some code from my undergrad, although I'm not too sure about parameter values. These are basically equations of motion, with $\phi$ and $\Delta E$ instead of $x$ and $v$

$$
\begin{cases}
    \phi_{n+1} = \phi_n + a_0 \Delta E_n
\\  \Delta E_{n+1} = \Delta E_n + a_1 \left(\sin \phi_{n+1} - \sin{\phi_s} \right)
\end{cases}
$$
where
$$
a_0 = { 2\pi h \eta \over \beta^2 E} \quad a_1 = eV
$$

````python
import matplotlib.pyplot as plt
import numpy as np

def accelerator_EOM(phase, ΔE, params:list):
    a0, a1, phi_s = params
    dphase = a0*ΔE
    dE = a1*(np.sin(phase+dphase) - np.sin(phi_s))
    return dphase, dE

params = [1, 1, 0] # a0 a1 phi_s
linspace_phi = np.linspace(-2*np.pi, 2*np.pi, 20)
linspace_ΔE = np.linspace(-1, 1, 20)

phase, ΔE = np.meshgrid(linspace_phi, linspace_ΔE) 

def get_points(X, V, func, params):
    dx, dv =  np.zeros(X.shape), np.zeros(V.shape)

    xlim, ylim = X.shape
    for i in range(xlim):
        for j in range(ylim):
            xloc = X[i,j]
            vloc = V[i,j]
            dx[i,j], dv[i,j] = func(xloc, vloc, params)
    
    return(dx, dv)


def make_phase_space(X, V, dx, dv):
    fig = plt.figure(figsize = (12,4))
    plt.subplot(1,2,1)
    plt.quiver(X,V,dx,dv)
    plt.xlabel("x")
    plt.ylabel("v")
    plt.subplot(1,2,2)
    plt.streamplot(X,V,dx,dv)
    plt.xlabel("x")
    plt.ylabel("v")
    plt.suptitle("Phase Space")
    plt.show()

dPhase, dΔE = get_points(phase, ΔE, accelerator_EOM, params)
make_phase_space(phase, ΔE, dPhase, dΔE)
````

````python
tao.plot('beam_plot')
fig = plt.gcf()
axes = fig.get_axes()

def update(frame):
    tao.cmd(f'set curve T::beam_plot.x.c1 ix_ele_ref = {frame}')
    
    for ax in axes: # prevents drawing from being kept
        ax.clear()
    tao.plot('beam_plot', axes=axes)
    
    fig.canvas.draw() 
    return axes

ani = animation.FuncAnimation(fig, update, frames=range(0,1000,5), interval=50, blit=False)
ani.save('beam_animation.gif', writer='pillow')
````
